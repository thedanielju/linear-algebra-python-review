# vector scaling, dot products, geometry
scale a vector by a number, compute dot products manually, explain why dot products return scalars
understand norm as length, how vectors encode geometry

# python
no numpy - using * with numbers and list, accumulating values, math library

scaling means: multiply every component by the same number

```python
# scalar
def scale_vector(v, scalar):
    result = []
    for i in range(len(v)):
       result.append(v[i] * scalar)
    return result
```

```python
scale_vector([1, 5, 99, 3, 5], 10)
```

```python
# trap list multiplication
[1, 2, 3, 1, 2, 3] * 5
```

# dot products
first operation that changes dimensionality - accumulation function
input = 2 vectors; output = 1 number


```python
def dot_product(v1, v2):
    if len(v1) != len(v2):
        return
    result = 0
    for i in range(len(v1)):
        result += (v1[i]*v2[i])
    return result
```

```python
dot_product([1, 3, 5], [8, 3, 10])
```

# dot products answer a geometric question:
how much do these vectors point in the same direction?
    a. large positive result = strongly aligned
    b. zero = unrelated
    c. negative = opposed

used in ML for similarity search, attention scores, linear classifiers

# euclidean norm
square each component, sum them, take the square root

resembles the pythagorean theorem, generalized to n dimensions
    this is because hypotenuse of 2D vectors is literally the vector length!

squaring matters to get rid of negative lengths - removes sign while
preserving magnitude

norm is intended to measure how far from the origin the vector is

```python
import math

# note it has to be the same list!

def vector_norm(v):
    return math.sqrt(dot_product(v, v))

vector_norm([3, 4])
```
