```python
# given two vectors in ℝⁿ, produce a new vector 
# w/ components sum of the inputs' components
# real vector addition requires a function
def add_vectors(v1, v2):
    if len(v1) != len(v2):
        return
    result = []
    for i in range(len(v1)): # produce a sequence of integers starting at 0 ... n
        result.append(v1[i] + v2[i])
    return result
```
