and how they differ from python lists in ways ML cares about

```python
import numpy as np

v_list = [1, 2, 3]
v_np = np.array([1, 2, 3])
```

```python
type(v_list)
```

```python
type(v_np)
```

```python
print(v_np)
print(v_np.shape)
print(v_np.dtype)
```

# basic numpy
shape tells us this is a 1D array with 3 elements
comma MATTERS! (3,) is a 1-tuple
    tuples: sequence type of pairs, described array's dimensions

dtype tells us every element is stored as a fixed size 64 bit integer
python lists store references to full objects

```python
print(v_list * 2)
print(v_np * 2)
```

```python
u_np = np.array([10, 20, 30])
print(v_np + u_np)

u_list = [10, 20, 30]
print(v_list + u_list)

# reminder - python lists use sequence rules and concatenate
# numpy arrays use vector math rules (elementwise)
```

```python
x = np.array([1, 2, 3])
y = x / 2
print(y)
print(y.dtype)
```

```python
x32 = np.array([1, 2, 3], dtype=np.float32)
print(x32.dtype)
print((x32 / 2).dtype)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a * b) # element wise - attention mechanisms
print(a @ b) # dot products
print(np.dot(a, b)) # dot products
```

```python
a = np.array([1, 2, 3])
print(a.shape)

s = a @ a
print(s, type(s))
```

# technically, the dot product is |u||v|cos(theta)
cos(theta) indicates properties

we normalize (scale control) because otherwise, dot products
are not useful indicators of directional/angular similarity
since magnitude would dominate, not direction. training
can crank magnitudes up and explode our similarity scores

# unit vector = vector with norm 1


```python
# normalizing unit vector
uv = v_np/np.linalg.norm(v_np)
print(uv)
print(np.linalg.norm(uv))
```
